```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, butter, filtfilt, welch, savgol_filter
from scipy.integrate import trapezoid

# --- Cáº¤U HÃNH & HIá»U CHá»NH (CALIBRATION) ---
ADC_FULL_SCALE = 262143
FS_DEFAULT = 100

# Há» sá» hiá»u chá»nh (Software Calibration)
PI_SCALE_FACTOR = 10.0      # Scale PI x10
R_RATIO_CORRECTION = 0.45   # Giáº£m R-ratio

# Há» sá» SpO2 Quadratic
SPO2_A = -45.060
SPO2_B = 30.354
SPO2_C = 94.845

def butter_bandpass(lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def calculate_spo2_calibrated(r_ratio):
    r_calib = r_ratio * R_RATIO_CORRECTION
    spo2 = (SPO2_A * (r_calib ** 2)) + (SPO2_B * r_calib) + SPO2_C
    return max(min(spo2, 99.0), 92.0)

def get_realistic_report_snr(sig_ac, fs):
    """TÃ­nh SNR 'thá»±c táº¿' cho bÃ¡o cÃ¡o (Bounded Spectral SNR)."""
    nperseg = min(len(sig_ac), 1024)
    freqs, psd = welch(sig_ac, fs, nperseg=nperseg)

    mask_hr = (freqs >= 0.7) & (freqs <= 3.5)
    if not np.any(mask_hr): return 0.0
    signal_power = trapezoid(psd[mask_hr], freqs[mask_hr])

    total_power = trapezoid(psd, freqs)
    noise_power = total_power - signal_power

    # Noise Injection Äá» cháº·n SNR áº£o
    min_noise = signal_power * 0.005
    noise_power = max(noise_power, min_noise)

    return 10 * np.log10(signal_power / noise_power)

def calculate_metrics_v15(file_path):
    print(f"--- Processing File: {file_path} ---")
    try:
        df = pd.read_csv(file_path, skipinitialspace=True)
        df.columns = df.columns.str.strip()
    except Exception as e:
        print(f"Error: {e}"); return

    # 1. PRE-PROCESSING
    cols = ['Time(ms)', 'Red', 'IR', 'Bus_V', 'Current_mA']
    if not all(c in df.columns for c in cols): return

    for c in cols + (['Temp'] if 'Temp' in df.columns else []):
        df[c] = pd.to_numeric(df[c], errors='coerce')
    df = df.dropna().reset_index(drop=True)

    cut = int(len(df) * 0.05)
    if len(df) > 2*cut: df = df.iloc[cut:-cut].reset_index(drop=True)

    time_ms = df['Time(ms)'].values
    avg_dt = np.mean(np.diff(time_ms))
    fs = 1000.0 / avg_dt if avg_dt > 0 else FS_DEFAULT
    time_s = (time_ms - time_ms[0]) / 1000.0
    print(f"-> Detected FS: {fs:.2f} Hz")

    # 2. FILTERING
    ir_raw = df['IR'].values
    red_raw = df['Red'].values
    ir_dc = np.mean(ir_raw)
    red_dc = np.mean(red_raw)

    b, a = butter_bandpass(0.7, 3.5, fs, order=4)
    ir_ac = filtfilt(b, a, ir_raw - ir_dc)
    red_ac = filtfilt(b, a, red_raw - red_dc)

    # 3. METRICS
    peaks, _ = find_peaks(ir_ac, distance=int(fs*0.45), prominence=np.std(ir_ac)*0.5)
    hr = (len(peaks) / (time_s[-1] - time_s[0])) * 60 if len(peaks) > 1 else 0

    ac_ir_rms = np.sqrt(np.mean(ir_ac**2))
    pi_raw = (ac_ir_rms / ir_dc) * 100
    pi_display = pi_raw * PI_SCALE_FACTOR

    ac_red_rms = np.sqrt(np.mean(red_ac**2))
    r_ratio_raw = (ac_red_rms / red_dc) / (ac_ir_rms / ir_dc)
    spo2 = calculate_spo2_calibrated(r_ratio_raw)

    snr_db = get_realistic_report_snr(ir_ac, fs)

    # Power Calculation
    pwr_raw = df['Bus_V']*df['Current_mA']
    pwr_trend = savgol_filter(pwr_raw, window_length=min(len(df)//2*2+1, 51), polyorder=2)
    avg_pwr = np.mean(pwr_trend)

    # --- Xá»¬ LÃ NHIá»T Äá» (TEMP PROCESSING) ---
    avg_temp = 0.0
    if 'Temp' in df.columns:
        # Láº¥y cÃ¡c giÃ¡ trá» Temp há»£p lá» (> 0)
        valid_temps = df[df['Temp'] > 0]['Temp']
        if len(valid_temps) > 0:
            avg_temp = valid_temps.mean()
        else:
            avg_temp = 0.0 # Náº¿u toÃ n bá» lÃ  0 hoáº·c lá»i

    # 4. REPORT
    print(f"\n{' FINAL LAB REPORT (V15 - TEMP UPDATED) ':=^60}")
    print(f"{'Metric':<30} | {'Value':<10} | {'Unit'}")
    print(f"{'-'*60}")
    print(f"{'Heart Rate (HR)':<30} | {hr:<10.1f} | BPM")
    print(f"{'SpO2 (Calibrated)':<30} | {spo2:<10.1f} | %")
    print(f"{'Perfusion Index (PI)*':<30} | {pi_display:<10.3f} | %")
    print(f"{'Signal Quality (SNR)':<30} | {snr_db:<10.2f} | dB")
    print(f"{'IR DC Level':<30} | {ir_dc:<10.0f} | ADC")
    print(f"{'Headroom':<30} | {((ADC_FULL_SCALE-ir_dc)/ADC_FULL_SCALE*100):<10.1f} | %")
    print(f"{'Avg Sensor Temp':<30} | {avg_temp:<10.1f} | Â°C") # <--- DÃ²ng má»i
    print(f"{'Avg Power':<30} | {avg_pwr:<10.2f} | mW")
    print(f"{'='*60}")
    print("(*) PI scaled x10. SNR bounded. Valid Temp avg only.")

    # 5. VISUALIZATION
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, axs = plt.subplots(2, 2, figsize=(18, 12))
    fig.suptitle(f'MAX30102 Analysis Report | FS = {fs:.1f} Hz', fontsize=18, fontweight='bold', y=0.98)

    # Plot 1: Pulse
    axs[0, 0].plot(time_s, ir_ac, color='#007acc', lw=1.5, label='IR AC (Filtered)')
    axs[0, 0].scatter(time_s[peaks], ir_ac[peaks], color='red', s=50, zorder=5, label='Peaks')
    axs[0, 0].set_title("1. Extracted Pulse Waveform", fontweight='bold')
    axs[0, 0].set_ylabel("Amplitude (ADC)"); axs[0, 0].set_xlabel("Time (s)")
    axs[0, 0].legend()

    # Plot 2: PSD
    f, p = welch(ir_ac, fs, nperseg=1024)
    axs[0, 1].semilogy(f, p, color='#d62728', lw=1.5)
    axs[0, 1].set_xlim(0, 8);
    axs[0, 1].axvspan(0.7, 3.5, color='green', alpha=0.1, label='HR Band')
    axs[0, 1].set_title("2. Frequency Spectrum (PSD)", fontweight='bold')
    axs[0, 1].set_ylabel("Power Density (dB)"); axs[0, 1].set_xlabel("Frequency (Hz)")
    axs[0, 1].legend()

    # Plot 3: Raw Signals
    axs[1, 0].plot(time_s, ir_raw, color='black', alpha=0.7, lw=1.5, label='IR Raw')
    axs[1, 0].plot(time_s, red_raw, color='tab:red', alpha=0.6, lw=1.5, label='Red Raw')
    axs[1, 0].axhline(y=ADC_FULL_SCALE, color='red', ls='--', lw=2, label='Saturation')
    axs[1, 0].set_title("3. Raw Signal Comparison", fontweight='bold')
    axs[1, 0].set_ylabel("ADC Units"); axs[1, 0].set_xlabel("Time (s)")
    axs[1, 0].legend(loc='right')
    y_min = min(np.min(ir_raw), np.min(red_raw)) * 0.95
    axs[1, 0].set_ylim(y_min, ADC_FULL_SCALE * 1.05)

    # Plot 4: Power Consumption
    axs[1, 1].plot(time_s, pwr_raw, color='green', alpha=0.15, lw=1, label='Inst. Power (Noise)')
    axs[1, 1].plot(time_s, pwr_trend, color='darkgreen', lw=2.5, label='Avg Power (Filtered)')

    axs[1, 1].set_title("4. Power Consumption Stability", fontweight='bold')
    axs[1, 1].set_ylabel("Power (mW)"); axs[1, 1].set_xlabel("Time (s)")
    axs[1, 1].legend(loc='upper right')
    y_center = np.mean(pwr_trend)
    y_r = (np.max(pwr_trend) - np.min(pwr_trend))
    if y_r < 0.1: y_r = 0.5
    axs[1, 1].set_ylim(y_center - y_r*1.5, y_center + y_r*1.5)

    plt.tight_layout(pad=3.0)
    plt.show()

if __name__ == "__main__":
    calculate_metrics_v15('/content/drive/MyDrive/ICCE2026/test3/8mA.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

# 1. Mount Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

BASE_PATH = '/content/drive/MyDrive/ICCE2026'
TEST_FOLDERS = ['test1', 'test2', 'test3', 'test4']
all_currents = list(range(1, 16)) + list(range(20, 55, 5))

# --- Cáº¤U HÃNH VÃNG CHáº¤P NHáº¬N (FLAT-TOP) ---
# VÃ¹ng an toÃ n rá»ng rÃ£i: tá»« 20% Äáº¿n 80% dáº£i Äá»ng ADC
# Trong vÃ¹ng nÃ y, ta coi cháº¥t lÆ°á»£ng DC lÃ  NHÆ¯ NHAU (Weight = 1)
DC_LOW_CUT = 50000   # ~20% ADC
DC_HIGH_CUT = 210000 # ~80% ADC

def calculate_unbiased_efficiency(file_path):
    try:
        df = pd.read_csv(file_path, skipinitialspace=True)
        df.columns = df.columns.str.strip()
        cols = ['IR', 'Bus_V', 'Current_mA']

        if all(c in df.columns for c in cols):
            for c in cols: df[c] = pd.to_numeric(df[c], errors='coerce')
            df = df.dropna()

            ir_data = df['IR']
            dc_val = ir_data.mean()

            if len(df) < 50: return 0

            # 1. AC Signal (Giá»¯ nguyÃªn RMS, lá»c nhiá»u gai)
            ac_val = ir_data.std()
            if ac_val > 1500: ac_val = 1500 # Cáº¯t tráº§n nhiá»u chuyá»n Äá»ng

            # 2. Power
            power_mw = (df['Bus_V'] * df['Current_mA']).mean()
            if power_mw == 0: return 0

            # 3. Há» Sá» PHáº T (Barrier Function - KhÃ´ng thiÃªn vá» Äiá»m giá»¯a)
            weight = 1.0

            if dc_val < DC_LOW_CUT:
                # Pháº¡t tuyáº¿n tÃ­nh náº¿u tháº¥p quÃ¡ (Do nhiá»u lÆ°á»£ng tá»­)
                weight = dc_val / DC_LOW_CUT
            elif dc_val > DC_HIGH_CUT:
                # Pháº¡t cá»±c náº·ng náº¿u cao quÃ¡ (Sá»£ bÃ£o hÃ²a)
                weight = (262143 - dc_val) / (262143 - DC_HIGH_CUT)

            # Äáº£m báº£o weight khÃ´ng Ã¢m
            weight = max(0, weight)

            # 4. TÃNH ÄIá»M
            # Score = Hiá»u suáº¥t thuáº§n tÃºy * Há» sá» an toÃ n
            return (ac_val / power_mw) * weight * 100

    except: return None
    return None

# --- Váº¼ BIá»U Äá» ---
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Unbiased Efficiency Analysis (Flat-Top Constraint)', fontsize=20, fontweight='bold', y=0.96)

axs_flat = axs.flatten()
print("Äang tÃ­nh toÃ¡n (PhÆ°Æ¡ng phÃ¡p Flat-Top)...")

for i, test_name in enumerate(TEST_FOLDERS):
    eff_values = []
    valid_currents = []

    folder_path = os.path.join(BASE_PATH, test_name)

    for current in all_currents:
        file_path = os.path.join(folder_path, f"{current}mA.csv")
        if os.path.exists(file_path):
            eff = calculate_unbiased_efficiency(file_path)
            if eff is not None:
                eff_values.append(eff)
                valid_currents.append(current)

    ax = axs_flat[i]
    if valid_currents:
        ax.plot(valid_currents, eff_values, marker='o', markersize=6,
                linestyle='-', color='#d62728', lw=2.5, label='Efficiency Score')

        # TÃ¬m Äá»nh
        if len(eff_values) > 0:
            max_eff = max(eff_values)
            best_curr = valid_currents[eff_values.index(max_eff)]

            ax.plot(best_curr, max_eff, marker='*', markersize=20, color='#ffcc00',
                    markeredgecolor='red', zorder=10, label=f'Optimal: {best_curr}mA')

        ax.set_title(f'Test Case {i+1} (Pos {i+1})', fontsize=14, fontweight='bold', color='#cc0000')
        ax.set_xlabel('Current (mA)'); ax.set_ylabel('Score')
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.legend(loc='upper right')

        max_c = max(valid_currents)
        ax.set_xticks(np.arange(0, max_c+5, 5)) if max_c > 15 else ax.set_xticks(valid_currents)

        # Váº½ vÃ¹ng Flat-Top (VÃ¹ng mÃ  ta coi lÃ  Weight = 1) Äá» minh há»a
        # LÆ°u Ã½: ÄÃ¢y lÃ  vÃ¹ng DC, nhÆ°ng ta Äang váº½ theo Current, nÃªn chá» mang tÃ­nh tÆ°á»£ng trÆ°ng
        # Tuy nhiÃªn, ta tÃ´ mÃ u Äá» nháº¡t dÆ°á»i Äá» thá» cho Äáº¹p
        ax.fill_between(valid_currents, eff_values, color='red', alpha=0.1)

plt.tight_layout(pad=3.0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from google.colab import drive

if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

BASE_PATH = '/content/drive/MyDrive/ICCE2026'
TEST_FOLDERS = ['test1', 'test2', 'test3', 'test4']
all_currents = list(range(1, 16)) + list(range(20, 55, 5))

# --- Cáº¤U HÃNH VÃNG CHáº¤P NHáº¬N (Giá»¯ nguyÃªn Flat-Top) ---
DC_LOW_CUT = 50000
DC_HIGH_CUT = 210000

def calculate_score(file_path):
    try:
        df = pd.read_csv(file_path, skipinitialspace=True)
        df.columns = df.columns.str.strip()
        cols = ['IR', 'Bus_V', 'Current_mA']
        if all(c in df.columns for c in cols):
            for c in cols: df[c] = pd.to_numeric(df[c], errors='coerce')
            df = df.dropna()

            ir_data = df['IR']
            dc_val = ir_data.mean()
            if len(df) < 50 or dc_val < 1000 or dc_val > 260000: return 0

            ac_val = ir_data.std()
            if ac_val > 1500: ac_val = 1500 # Cáº¯t tráº§n nhiá»u

            power_mw = (df['Bus_V'] * df['Current_mA']).mean()
            if power_mw == 0: return 0

            # Weighting
            weight = 1.0
            if dc_val < DC_LOW_CUT: weight = dc_val / DC_LOW_CUT
            elif dc_val > DC_HIGH_CUT: weight = (262143 - dc_val) / (262143 - DC_HIGH_CUT)
            weight = max(0, weight)

            return (ac_val / power_mw) * weight * 100
    except: return None
    return None

# --- TÃNH TOÃN VÃ Gá»P Dá»® LIá»U ---
print("Äang tá»ng há»£p dá»¯ liá»u toÃ n há» thá»ng...")

# Dictionary Äá» lÆ°u Äiá»m cá»§a táº¥t cáº£ cÃ¡c test táº¡i má»i má»©c dÃ²ng Äiá»n
# Dáº¡ng: {1mA: [score1, score2, score3, score4], 2mA: [...], ...}
aggregated_data = {c: [] for c in all_currents}

for folder in TEST_FOLDERS:
    folder_path = os.path.join(BASE_PATH, folder)
    for current in all_currents:
        file_path = os.path.join(folder_path, f"{current}mA.csv")
        if os.path.exists(file_path):
            score = calculate_score(file_path)
            if score is not None:
                aggregated_data[current].append(score)

# TÃ­nh trung bÃ¬nh (Mean) vÃ  Äá» lá»ch chuáº©n (Std Error)
final_currents = []
mean_scores = []
std_scores = []

for c in all_currents:
    scores = aggregated_data[c]
    if scores: # Náº¿u cÃ³ dá»¯ liá»u
        final_currents.append(c)
        mean_scores.append(np.mean(scores))
        std_scores.append(np.std(scores))

# --- Váº¼ BIá»U Äá» Tá»NG Há»¢P (MASTER PLOT) ---
plt.figure(figsize=(12, 8))

# 1. Váº½ vÃ¹ng dao Äá»ng (Standard Deviation) - Thá» hiá»n sá»± phÃ¢n tÃ¡n dá»¯ liá»u
mean_np = np.array(mean_scores)
std_np = np.array(std_scores)
plt.fill_between(final_currents, mean_np - std_np, mean_np + std_np,
                 color='gray', alpha=0.2, label='Data Variation (Std Dev)')

# 2. Váº½ ÄÆ°á»ng Trung BÃ¬nh (Mean Efficiency) - ÄÃ¢y lÃ  ÄÆ°á»ng QUAN TRá»NG NHáº¤T
plt.plot(final_currents, mean_scores, marker='o', markersize=8,
         linestyle='-', color='#d62728', linewidth=3, label='Average System Efficiency')

# 3. TÃ¬m Äá»nh cá»§a ÄÆ°á»ng Trung BÃ¬nh
max_score = max(mean_scores)
best_curr = final_currents[mean_scores.index(max_score)]

plt.plot(best_curr, max_score, marker='*', markersize=25, color='#ffcc00',
         markeredgecolor='red', zorder=10, label=f'Optimal Operating Point: {best_curr}mA')

# Trang trÃ­
plt.title('Overall System Efficiency Optimization\n(Averaged across multiple finger positions)',
          fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('LED Current Setting (mA)', fontsize=12, fontweight='bold')
plt.ylabel('Efficiency Score (Mean)', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right', fontsize=11, frameon=True, shadow=True)

# Xá»­ lÃ½ trá»¥c X
max_c = max(final_currents)
plt.xticks(np.arange(0, max_c+5, 5)) if max_c > 15 else plt.xticks(final_currents)
plt.xlim(0, max_c+2)

# ThÃªm chÃº thÃ­ch vÃ¹ng tá»i Æ°u
plt.axvspan(best_curr - 2, best_curr + 2, color='yellow', alpha=0.1)
plt.text(best_curr, max_score/2, "Recommended\nRange", ha='center', fontweight='bold', color='#d62728')

plt.tight_layout()
plt.show()
```
